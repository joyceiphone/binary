1. First Implementation

```python
#original Code
def bsearch1(arr, key):
    low, high = 0, len(arr)
    while high - low > 1:
        mid = (low + high) // 2
        if arr[mid] == key:
            return mid
        elif arr[mid] < key:
            low = mid
        else:
            high = mid
    return None
```

```python
# changed Code
def bsearch1(arr, key):
    low, high = 0, len(arr)-1
    while high - low >= 0:
        mid = (low + high)//2
        if arr[mid] == key:
            return mid
        elif arr[mid] < key:
            low = mid + 1
        else:
            high = mid-1
    return None
```

1. The code does not return the index of the target value if the target is the first item of the array. The original code does not include the case when low = high.
2. Bugs are fixed on the original code below. 
3. Natural Test arr = [1, 2, 3, 4, 5], key = 1

Second Implementation

```python
#original code
def bsearch2(arr, key, left=0, right=None):
    if right is None:
        right = len(arr)
    if right < left:
        return None
    middle = (left + right) >> 1
    if arr[middle] > key:
        return bsearch2(arr, key, left, middle)
    if arr[middle] < key:
        return bsearch2(arr, key, middle + 1, right)
    return middle
```

```python
#fixed code
def bsearch2(arr, key, left=0, right=None):
    if right is None:
        right = len(arr)-1
    if right < left:
        return None
    middle = (left + right) >> 1
    if arr[middle] > key:
        return bsearch2(arr, key, left, middle - 1)
    elif arr[middle] < key:
        return bsearch2(arr, key, middle + 1, right)
    return middle
```

1. The code runs into infinite loop if the target is not present in the array. The original code keeps running as long as left >= right. The code will keep running when left == right(middle). Also we need to prevent the case when the target value is larger than the largest element in the array. We need to make the length of the search to be len(arr) - 1.
2. Fixed code is presented below.
3. Natural Test arr = [1, 2, 3, 4, 5], key = -1, arr = [1, 2, 3, 4, 5], key = 7

Third Implementation

```python
#original Code
def bsearch4(arr, key):
    n = len(arr)
    if n < 2:
        return (0 if (n == 1 and arr[0] == key) else None)
    m = int(0.5 * n)
    if arr[m] > key:
        return bsearch3(arr[:m], key)
    result = bsearch3(arr[m:], key)
    print(result)
    return (result + m if result != None else None)
```

```python
#fixed code
def bsearch3(arr, key, left = 0, right = None):
    if right == None:
        right = len(arr)
    if left >= right:
        return None
    mid = (left + right)//2
    if arr[mid] == key:
        return mid
    if arr[mid] > key:
        return bsearch3(arr, key, left, mid)
    return bsearch3(arr, key, mid + 1, right)
```

1. The original code has time complexity $n\log(n)$, because slicing has complexisty $O(N)$ each time if cut the array into half.  
2. Fixed code is under the original code. We add left and right.
3. The details of showing why the complexisy is $n\log(n)$ is below. 

```python
import matplotlib.pyplot as plt
import random
import time
```

```python
def generateRandomList(n):
    arr = []
    for i in range(n):
        arr.append(random.random())
    return arr
```

```python
def getArr2():
    arr1 = []
    arr2 = []
    
    sumBinaryTime = 0
    sumSlicingTime = 0
    
    for i in range(0, 10000):
        numList = generateRandomList(i)
        x = i*random.random()
        
        sbt = time.time()
        binaryFound = bsearch3(numList, x)
        ebt = time.time()
        elapsedBinaryTime = ebt - sbt
        
        slt = time.time()
        slicingFound = bsearch4(numList, x)
        elt = time.time()
        elapsedSlicingTime = elt - slt
        
        sumBinaryTime += elapsedBinaryTime
        sumSlicingTime += elapsedSlicingTime
        
        arr1.append(sumBinaryTime)
        arr2.append(sumSlicingTime)
    
    return arr1,arr2
```

```python
arr1, arr2 = getArr2()
```

```python
plt.loglog(arr1, 'r', label = "binarySearch")
plt.loglog(arr2, 'g', label = "slicingSearch")
plt.legend()
```
